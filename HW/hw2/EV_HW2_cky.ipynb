```python
'''
Homework assign_2 PRNG'S (Psuedo Random Numeric in Python)
'''
import random
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter

'''
PART I
'''

seq1= np.random.uniform(0.0, 1.0, size=10000)
#print(seq1)
#print(len(seq1))

n, bins, patches = plt.hist(seq1, 20, color='#75BEBB')
plt.title("Histogram of  uniform distribution in interval [0.0:1.0]")
plt.xlabel("Values")
plt.ylabel("Times")
plt.savefig('uniform.png')
plt.show()

'''
PART II Gaussian Distribution with Historgram
'''
mean = 5.0
stddev = 2.0
seq2 = np.random.normal(mean,stddev,10000)      # mu(i.e.mean) = 5.0 and sigma (i.e.stddev = standard deviation) = 2.0
#print(len(seq2))
n,bins,patches = plt.hist(seq2, 50,density=True)
plt.plot(bins, 1/(stddev * np.sqrt(2 * np.pi)) *np.exp( - (bins - mean)**2 / (2 * stddev**2) ),linewidth=2, color='r')
plt.title("Gaussian Distrbution with mean 5.0 and std 2.0")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.savefig('Gaussian5_2.png')
plt.show()


'''
PART III Dice_roll_Generator
'''
# dice-roll generator
def dice_roll_generator(num_of_dice:int , size:int , num_of_trials:int):
    sum_list = []
    for _ in range(num_of_trials):
        dice_sum = 0
        for _ in range(num_of_dice):
            dice_sum+= random.randint(1,size)
            
        sum_list.append(dice_sum)
         
    return sum_list   

trials = 10000
#2d6 (Two dice rolls , 6 - sided , sum the 2 dicce )
list_of_2d6 = []
list_of_2d6 = dice_roll_generator(2,6,trials)
plt.hist(list_of_2d6,50)
plt.title('Sum_of_2d6')
plt.xlabel('Sum')
plt.ylabel('Times')
plt.savefig('2d6.png')
plt.show()

#1d12
list_of_1d12 = []
list_of_1d12 = dice_roll_generator(1,12,trials)
plt.hist(list_of_1d12,50)
plt.title('Sum_of_1d12')
plt.xlabel('Sum')
plt.ylabel('Times')
plt.savefig('1d12.png')
plt.show()

#2d10
list_of_2d10 = []
list_of_2d10 = dice_roll_generator(2,10,trials)
plt.hist(list_of_2d10,50)
plt.title('Sum_of_2d10')
plt.xlabel('Sum')
plt.ylabel('Times')
plt.savefig('2d10.png')
plt.show()

#1d20
list_of_1d20 = []
list_of_1d20 = dice_roll_generator(1,20,trials)
plt.hist(list_of_1d20,50)
plt.title('Sum_of_1d20')
plt.xlabel('Sum')
plt.ylabel('Times')
plt.savefig('1d20.png')
plt.show()


```
